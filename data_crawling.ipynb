```python
import pandas as pd
```

```python
y_data = {year: [] for year in range(2018, 2024)}

for y in range(2018, 2024):
        url = f'https://www.kleague.com/record/player.do?year={y}'
        try:
            dfs = pd.read_html(url)
            for df in dfs:
                df['Year'] = y
                y_data[y].append(df)
        except Exception as e:
            print(f"Error reading {url}: {e}")
```

```python
for year, tables in y_data.items():
    if tables:
        year_data = pd.concat(tables, ignore_index=True)
        year_data.to_csv(f'ranking_{year}.csv', index=False)
        print(f'Saved data for year {year}')
    else:
        print(f'No data available for year {year}')
```

```python
import os
from dotenv import load_dotenv
load_dotenv(verbose = True)
path = os.getenv('url')
r_2018 = pd.read_csv(os.path.join(path, 'ranking_2018.csv'))
r_2019 = pd.read_csv(os.path.join(path, 'ranking_2019.csv'))
r_2020 = pd.read_csv(os.path.join(path, 'ranking_2020.csv'))
r_2021 = pd.read_csv(os.path.join(path, 'ranking_2021.csv'))
r_2022 = pd.read_csv(os.path.join(path, 'ranking_2022.csv'))
r_2023 = pd.read_csv(os.path.join(path, 'ranking_2023.csv'))
r_2021
```

```python
total = pd.concat([r_2018, r_2019, r_2020, r_2021, r_2022, r_2023])
total
```

```python
total = total.drop(['순위', '선수명', '팀', '무실점경기', 'Year', '코너킥', '파울', '실점', '오프사이드', '경고', '퇴장', '경기당 기록'], axis = 1)
total
```

```python
total.to_csv('total_ranking.csv', index = False)
```

```python
t = pd.read_csv(os.path.join(path, 'total_ranking.csv'))
t
```

```python
from sklearn.cluster import KMeans
ks = range(1, 15)
inertias = []
for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(total)
    inertias.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20, 20))
plt.plot(ks, inertias)
plt.xticks(ks)
plt.show()
```

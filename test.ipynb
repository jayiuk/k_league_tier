```python
import pandas as pd
import pickle

import os
from dotenv import load_dotenv
load_dotenv(verbose = True)
path = os.getenv('url')
data = pd.read_csv(os.path.join(path, 'total_ranking.csv'))

with open('tier_model', 'rb') as f:
    kmtest = pickle.load(f)
kmtest.fit(data)
pred = kmtest.predict(data)

    
c_data = data.copy()
c_data['티어'] = pred
c_data
```

```python
cluster = c_data.groupby('티어')
```

```python
cluster.mean()
```

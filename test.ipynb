```python
import pandas as pd
import pickle

import os
from dotenv import load_dotenv
import xgboost as xgb

load_dotenv(verbose = True)
path = os.getenv('url')
data = pd.read_csv(os.path.join(path, 'tier.csv'))
y_data0 = data['target']
x_data0 = data.drop(['target'], axis = 1)




dto = xgb.DMatrix(data = x_data0, label = y_data0)

with open('xgb_tier.model', 'rb') as f:
    xgbtest = pickle.load(f)
pred = xgbtest.predict(dto)

    
c_data = x_data0.copy()
c_data['티어'] = pred
c_data
```

```python
cluster = c_data.groupby('티어')
```

```python
cluster.mean()
```
